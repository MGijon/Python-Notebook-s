```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets
```

```python
bh = datasets.load_boston()   # loading the data set
X = bh.data
y = bh.target
```

First exploration of the dataset:

```python
fields = bh.feature_names
print(fields)
```

```python
plt.figure(1)
plt.subplot(211)
sns.lvplot(x = X[:, 0])
plt.subplot(212)
sns.distplot(X[:, 0])
plt.show()
```

```python
X_noExtremo = [j for j in X[:, 0] if j < 2]
plt.figure(1)
plt.subplot(211)
sns.lvplot(x = X_noExtremo)
plt.subplot(212)
sns.distplot(X_noExtremo)
plt.show()
```
